Ejercicio 1<br>

Escribe un programa que muestre por pantalla la concatenación de un número y una cadena
de caracteres. Para obtener esta concatenación puedes usar uno de los operadores explicados
en este tema. Ejemplo: dado el número 3 y la cadena ‘abc’, el programa mostrará la cadena
‘3abc’

```python
numero = 3
cadena = 'abc'

# Pasamos el número a string y concatenamos con +
resultado = str(numero) + cadena

print(resultado)  # Debería mostrar: 3abc
```

Ejercio 2<br>

Escribe un programa que muestre por pantalla un valor booleano que indique si un número
entero N está contenido en un intervalo semiabierto [a,b), el cual establece una cota inferior
a (inclusive) y una cota superior b (exclusive) para N.


```python
N = 5
a = 1
b = 10

# Intervalo semiabierto [a, b): a <= N < b
esta_en_intervalo = (N >= a) and (N < b)

print(esta_en_intervalo)
```

Ejercicio 3<br>

Escribe un programa que, dado dos strings S1 y S2 y dos números enteros N1 y N2, determine
si el substring que en S1 se extiende desde la posición N1 a la N2 (ambos inclusive) está
contenido en S2.


```python
S1 = "programacion"
S2 = "me gusta la programacion en python"

N1 = 0   # primera posición (recuerda que empieza en 0)
N2 = 10  # última posición (inclusive)

# En Python el slicing es [inicio:fin_exclusivo], por eso usamos N2 + 1
subcadena = S1[N1:N2+1]

esta_contenido = subcadena in S2

print("Subcadena:", subcadena)
print("¿Está contenida en S2?:", esta_contenido)
```

Ejercicio 4 <br>

Dada una lista con elementos duplicados, escribir un programa que muestre una nueva lista
con el mismo contenido que la primera pero sin elementos duplicados.

```python
lista = [1, 2, 2, 3, 4, 4, 4, 5]

lista_sin_duplicados = []
for elemento in lista:
    if elemento not in lista_sin_duplicados:
        lista_sin_duplicados.append(elemento)

print("Lista original:      ", lista)
print("Lista sin duplicados:", lista_sin_duplicados)
```

Ejercicio 5<br>

Escribe un programa que, dada una lista de strings L, un string s perteneciente a L y un
string t, reemplace s por t en L. El programa debe mostrar la lista resultante por pantalla.


```python
L = ["python", "java", "python", "c++"]
s = "python"
t = "py"

# Recorremos la lista y reemplazamos
for i in range(len(L)):
    if L[i] == s:
        L[i] = t

print(L)
```

Ejercicio 6<br>

Escribe un programa que defina una tupla con elementos numéricos, reemplace el valor del
último por un valor diferente y muestre la tupla por pantalla. Recuerda que las tuplas son
inmutables. Tendrás que usar objetos intermedios.

```python
tupla = (10, 20, 30, 40)
nuevo_valor_ultimo = 99

# Convertimos a lista
lista_temp = list(tupla)

# Modificamos el último elemento
lista_temp[-1] = nuevo_valor_ultimo

# Volvemos a tupla
tupla_modificada = tuple(lista_temp)

print("Tupla original:    ", tupla)
print("Tupla modificada:  ", tupla_modificada)
```

Ejercicio 7 <br>

Dada la lista [1,2,3,4,5,6,7,8] escribe un programa que, a partir de esta lista, obtenga la lista
[8,6,4,2] y la muestre por pantalla.


```python
lista = [1,2,3,4,5,6,7,8]

# Podemos tomar los pares empezando desde el final
resultado = lista[::-1]  # lista al revés: [8,7,6,5,4,3,2,1]

# Nos quedamos solo con los pares
resultado = [x for x in resultado if x % 2 == 0]

print(resultado)  # [8, 6, 4, 2]
```

Ejercicio 8 <br>

Escribe un programa que, dada una tupla y un índice válido i, elimine el elemento de la tupla
que se encuentra en la posición i. Para este ejercicio sólo puedes usar objetos de tipo tupla.
No puedes convetir la tupla a una lista, por ejemplo.

```python
tupla = (10, 20, 30, 40, 50)
i = 2  # eliminar el elemento en posición 2 (30)

# Partimos la tupla en dos y las juntamos sin el elemento i
tupla_sin_i = tupla[:i] + tupla[i+1:]

print("Tupla original:     ", tupla)
print("Índice eliminado:", i)
print("Tupla resultante:   ", tupla_sin_i)
```

Ejercicio 9 <br>

Escribe un programa que obtenga la mediana de una lista de números. Recuerda que la
mediana M de una lista de números L es el número que cumple la siguiente propiedad: la
mitad de los números de L son superiores a M y la otra mitad son inferiores. Cuando el
número de elementos de L es par, se puede considerar que hay dos medianas. No obstante,
en este ejercicio consideraremos que únicamente existe una mediana

```python
L = [7, 1, 3, 9, 5]

# Primero ordenamos la lista
L_ordenada = sorted(L)
n = len(L_ordenada)

if n % 2 == 1:
    # cantidad impar: un solo valor central
    mediana = L_ordenada[n // 2]
else:
    # cantidad par: promedio de los dos del centro
    medio1 = L_ordenada[n // 2 - 1]
    medio2 = L_ordenada[n // 2]
    mediana = (medio1 + medio2) / 2

print("Lista ordenada:", L_ordenada)
print("Mediana:", mediana)
```
